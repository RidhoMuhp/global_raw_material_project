```python
fossil_export_import_melted = fossil_export_import.melt(id_vars=['Flow name', 'Category'], value_vars=year_cols, var_name='Year', value_name='Value')
fossil_export_import_melted
```

```python
fossil_export_import_melted['Year'] = fossil_export_import_melted['Year'].astype(int)
fossil_export_import_melted['Value'] = fossil_export_import_melted['Value'].astype(float)
fossil_export_import_melted.info()
```

```python
fossil_export_melted = fossil_export.melt(id_vars=['Flow name', 'Category'], value_vars=year_cols, var_name='Year', value_name='Value')
fossil_export_melted['Year'] = fossil_export_melted['Year'].astype(int)
fossil_export_melted['Value'] = fossil_export_melted['Value'].astype(float)
fossil_export_melted.info()
```

```python
fossil_import_melted = fossil_import.melt(id_vars=['Flow name', 'Category'], value_vars=year_cols, var_name='Year', value_name='Value')
fossil_import_melted['Year'] = fossil_import_melted['Year'].astype(int)
fossil_import_melted['Value'] = fossil_import_melted['Value'].astype(float)
fossil_import_melted.info()
```

```python
plt.plot(fossil_export_melted['Year'], fossil_export_melted['Value'])
plt.title('Fossil Fuel Exports from Indonesia Over the Years')
plt.xlabel('Year')
plt.ylabel('Export Value')
plt.show()
```

```python
plt.plot(fossil_import_melted['Year'], fossil_import_melted['Value'])
plt.title('Fossil Fuel Imports to Indonesia Over the Years')
plt.xlabel('Year')
plt.ylabel('Import Value')  
plt.show()
```
